## Activation function

```python
import matplotlib.pylab as plt 
import numpy as np

from matplotlib import rcParams

# figure size in inches
rcParams['figure.figsize'] = 11.7,8.27 # A4 format

#  simulated biological neuron activation
```

$f(x) = \frac{1}{1+ \exp(-ax)}$

```python
x= np.arange(-2*np.pi,2*np.pi,0.1)
a = 1
f1 = 1/(1+np.exp(-a*x))
df = f1*(1-f1)

plt.figure(figsize=(11.7,8.27))
plt.plot(x,f1)
plt.plot(x,df)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
# connectic nodes: weights
w1 = 0.5; w2 = 1.0; w3 = np.pi;
l1 = 'w= 0.5'; l2 = 'w= 1.0'; l3 = 'w= pi'; 

plt.figure(figsize=(11.7,8.27))

for  w,l in [(w1,l1),(w2,l2),(w3,l3)]:
    f = 1 /(1+np.exp(-x*w))
    plt.plot(x,f,label =l)

plt.xlabel ('x')
plt.ylabel ('h_w(x)')
plt.legend(loc=2)
plt.show()
```

```python
# connectic nodes: bias
b1= -np.pi; b2= 0; b3= np.pi
l1 = 'b= -pi'; l2 = 'w= 0'; b3 = 'b= pi'; 

plt.figure(figsize=(11.7,8.27))
for  b,l in [(b1,l1),(b2,l2),(w3,l3)]:
    f = 1 /(1+np.exp(-(x*w+b)))
    plt.plot(x,f,label =l)

plt.xlabel ('x')
plt.ylabel ('h_w(x)')
plt.legend(loc=2)
plt.show()
```
